```python
from selenium import webdriver
from bs4 import BeautifulSoup
from Base_de_datos.basededatos import BaseDeDatos
import re
import time
import requests
from selenium.webdriver.common.by import By
import locale
locale.setlocale(locale.LC_ALL, 'es_ES.utf-8')
bd = BaseDeDatos()
```

```python
#Maximus gaming
#Estan las placas AMD y Nvidia en la misma pagina
url = 'https://www.maximus.com.ar'
bd = BaseDeDatos()
```

```python
driver = webdriver.Chrome()

driver.get(url)
elemento = driver.find_element(By.ID, 'placasDeVideo')
elemento.click()
html = driver.page_source
time.sleep(5)
driver.quit()
html = BeautifulSoup(html, 'html.parser')
productos = html.find_all('div',class_='product')
productos
```

```python
TIENDA =bd.idTienda("maximus gaming")

for producto in productos:    
    try:
        nombre = producto.find('div', class_='description').find('h4').find('a').text.strip()
        nombre = nombre[14:]
        
        link = producto.find('div', class_='description').find('h4').find('a')['href']

        if nombre.find('Rtx') != -1 or nombre.find('Nvidia') != -1 or nombre.find('Gtx') != -1 or nombre.find('Pny') != -1 or nombre.find("Gt") != -1:
            marca = 'Nvidia'
        else:
            marca = 'AMD'

        price_div = producto.find('div', class_='cajaprecio').find('div', class_='price')
        for small_tag in price_div.find_all('small'):
            small_tag.extract() 
            
        precio = price_div.text.strip()
        precio = re.sub(r'[^0-9.]', '', precio)
        precio = precio.replace(".","")
        precio = int(precio)

        url_imagen = producto.find('div', class_='image').find('a').find('div').find('img')['src']

        respuesta = requests.get(url_imagen)
        imagen = respuesta.content
        print(nombre[:14])
        print(link)
        print(TIENDA)
        print(precio)
        print(marca)
        print(imagen)
        bd.insertarPlaca(nombre=nombre,precio=precio,link=link,marca=marca,tienda=TIENDA,imagen=imagen)

        imagen = b''
        url_img = ''
        respuesta = b''


    except Exception as e:
        print(e)
driver.quit()
```
