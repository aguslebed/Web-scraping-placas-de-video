```python
from selenium import webdriver
from bs4 import BeautifulSoup
from Base_de_datos.basededatos import BaseDeDatos
import re
import time
import requests
from selenium.webdriver.common.by import By
import locale
locale.setlocale(locale.LC_ALL, 'es_ES.utf-8')
```

```python
locale.setlocale(locale.LC_ALL, 'es_ES.utf-8')
url = 'https://compragamer.com'
urlAMD = 'https://compragamer.com/?seccion=3&cate=62'
urlNVIDIA = 'https://compragamer.com/?seccion=3&cate=6'
```

```python
bd = BaseDeDatos()
bd.eliminarPlacasDeUnaTienda(tienda="compra gamer")
```

```python

driver = webdriver.Chrome()

driver.get(urlNVIDIA)

htmlNVIDIA = driver.page_source
time.sleep(5)
driver.get(urlAMD)

htmlAMD = driver.page_source

driver.quit()

htmlNVIDIA = BeautifulSoup(htmlNVIDIA, 'html.parser')

htmlAMD = BeautifulSoup(htmlAMD, 'html.parser')

productosNvidia = htmlNVIDIA.find('div', class_="col-12 col-lg-8 col-xl-8 col-md-7 noPaddingLeft paddingRightTablet paddingEspecial").find('div', class_="addRow ng-star-inserted").find_all('cgw-product-alone', class_='mat-card card-product ng-star-inserted')
productosAmd = htmlAMD.find('div', class_="col-12 col-lg-8 col-xl-8 col-md-7 noPaddingLeft paddingRightTablet paddingEspecial").find('div', class_="addRow ng-star-inserted").find_all('cgw-product-alone', class_='mat-card card-product ng-star-inserted')
fabricantes = [productosNvidia,productosAmd]
```

```python
productosAmd
```

```python
driver = webdriver.Chrome()
cont = 0
TIENDA = bd.idTienda("Compra Gamer")
marca = "Nvidia"
for fabricante in fabricantes:
    for producto in fabricante:
        try:
            time.sleep(30)
            nombre = producto.find('div', class_='contenidoPrincipal').find('a').find('span').text.strip()
            nombre  =  nombre[14:]

            link = url + producto.find('a', href=True)['href']
            precio = producto.find('div', class_='contenidoPrincipal').find('div', class_='contenedor-price').find('h1').find('span').text.strip()
            precio = re.sub(r'[^0-9.]', '', precio)
            precio = precio.replace(".","")
            precio = int(precio)


            #####################
            driver.get(link)
            html = driver.page_source
            html = BeautifulSoup(html, 'html.parser')
            try:
                url_img = html.find('div', class_='client-thumbnails').find('div', class_="slide slick-slide slick-current slick-active ng-star-inserted").find('img').get('src')
                respuesta = requests.get(url_img)
                imagen = respuesta.content
            except Exception as e:
                print(e)
            ####################
            

            
            
            cont += 1
            print("Placa nro: ",cont)
            bd.insertarPlaca(nombre=nombre,precio=precio,link=link,marca=marca,tienda=TIENDA,imagen=imagen)
            
            imagen = b''
            url_img = ''
            respuesta = b''
        except Exception as e:
            print(e)

    marca ='AMD'

driver.quit()
```
